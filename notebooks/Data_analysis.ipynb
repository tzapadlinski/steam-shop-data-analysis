```python
df = df[['AppID', 'Name', 'Release date', 'Estimated owners', 'Peak CCU',
       'Required age', 'Price', 'DLC count',
       'Supported languages', 'Full audio languages', 'Windows',
       'Mac', 'Linux', 'Metacritic score',
       'Positive', 'Negative', 'Achievements', 'Recommendations', 'Average playtime forever', 'Average playtime two weeks',
       'Median playtime forever', 'Median playtime two weeks', 'Developers',
       'Publishers', 'Categories', 'Genres','Tags']]
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
from itertools import chain

raw_categories = df['Categories'].dropna().astype(str)
category_lists = raw_categories.apply(lambda x: x.split(',') if x else [])
all_categories = list(chain.from_iterable(category_lists))
unique_categories = set(c.strip() for c in all_categories if c.strip())

raw_genres = df['Genres'].dropna().astype(str)
genre_lists = raw_genres.apply(lambda x: x.split(',') if x else [])
all_genres = list(chain.from_iterable(genre_lists))
unique_genres = set(g.strip() for g in all_genres if g.strip())

raw_tags = df['Tags'].dropna().astype(str)
tag_lists = raw_tags.apply(lambda x: x.split(',') if x else [])
all_tags = list(chain.from_iterable(tag_lists))
unique_tags = set(c.strip() for c in all_tags if c.strip())


print(f"Number of unique categories: {len(unique_categories)}")
print(f"Number of unique genres: {len(unique_genres)}")
print(f"Number of unique tags: {len(unique_tags)}")


print("\nCategories:", sorted(unique_categories))
print("\nGenres:", sorted(unique_genres))
```

### Halve classifier with most rows

```python
mask = df['Estimated owners'] == '0 - 20000'
filtered_df = df[mask]
sampled_df = filtered_df.sample(frac=0.8, random_state=42)
df = df.drop(sampled_df.index)
```

```python
cols = ['Price', 'Peak CCU', 'Positive', 'Negative']

plt.figure(figsize=(12, 8))

for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, i)
    
    df[col].hist(bins=50)
    plt.yscale('log')
    plt.title(f'{col} (log scale)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

columns_to_plot = [
    'Peak CCU', 'Price', 'DLC count', 'Metacritic score', 'Positive', 
    'Negative', 'Achievements', 'Recommendations', 'Average playtime forever', 
    'Median playtime forever', 'Months since release'
]

plt.figure(figsize=(12, 24))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(len(columns_to_plot), 1, i) 
    
    df[col].plot(kind='box', vert=False, patch_artist=True, 
                 boxprops=dict(facecolor='skyblue', color='black'),
                 whiskerprops=dict(color='black', linewidth=1.5),
                 flierprops=dict(markerfacecolor='red', marker='o', markersize=5))
    plt.yscale('log')
    plt.title(f'{col} - Boxplot', fontsize=12)
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
owner_mapping = {
    '0 - 0': 0,
    '0 - 20000': 1,
    '20000 - 50000': 2,
    '50000 - 100000': 3,
    '100000 - 200000': 4,
    '200000 - 500000': 5,
    '500000 - 1000000': 6,
    '1000000 - 2000000': 7,
    '2000000 - 5000000': 7,
    '5000000 - 10000000': 7,
    '10000000 - 20000000': 7,
    '20000000 - 50000000': 7,
    '50000000 - 100000000': 7,
    '100000000 - 200000000': 7
}

df['Estimated owners'] = df['Estimated owners'].map(owner_mapping)
```

```python
import matplotlib.pyplot as plt

owner_counts = df['Estimated owners'].value_counts()

plt.figure(figsize=(12, 6))
owner_counts.plot(kind='bar', color='skyblue')

plt.title('Distribution of Estimated Owners', fontsize=14)
plt.xlabel('Estimated Owners', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
